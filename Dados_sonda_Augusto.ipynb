```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator

# Carregando o arquivo
df_arq = "C:\\Users\\Carlos\\OneDrive\\FURG\\Last_YEAR\\Ecoveco_Fìsico-química água_Mensal_Total.xlsx"
dados = pd.read_excel(df_arq, sheet_name='Final')

# Transformando Ano, Mês e Dia em datetime e configurando como índice
dados['Data'] = pd.to_datetime(dados[['Ano', 'Mês', 'Dia']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d')
dados.set_index('Data', inplace=True)

# Definindo as datas dos Levantamentos Aéreos (exemplo)
levantamentos_aereos = [
    '2023-03-08',
    '2023-06-19','2023-11-16','2024-03-13','2024-04-25'
]
levantamentos_aereos = pd.to_datetime(levantamentos_aereos)

# Criando subplots para Temperatura, Salinidade e Turbidez
fig, axs = plt.subplots(3, 1, figsize=(14, 14), sharex=True)
locais = dados['Local'].unique()
colors = ['b', 'g', 'r']  # Azul, Verde e Vermelho para diferentes locais

# Função para adicionar pontos de destaque
def add_highlight_points(ax, data, y_column):
    for ponto_destaque in levantamentos_aereos:
        if ponto_destaque in data.index:
            valor_destaque = data.loc[ponto_destaque, y_column]
            ax.scatter(ponto_destaque, valor_destaque, color='k', s=100)

# Plot Temperatura
for local, color in zip(locais, colors):
    local_data = dados[dados['Local'] == local]
    axs[0].plot(local_data.index, local_data['Temperatura'], linestyle='-', label=local, color=color)
    add_highlight_points(axs[0], local_data, 'Temperatura')
axs[0].set_title('Temperatura')
axs[0].set_ylabel('Temperatura (°C)')
axs[0].legend()
axs[0].grid(True)

# Plot Salinidade
for local, color in zip(locais, colors):
    local_data = dados[dados['Local'] == local]
    axs[1].plot(local_data.index, local_data['Salinidade'], linestyle='-', label=local, color=color)
    add_highlight_points(axs[1], local_data, 'Salinidade')
axs[1].set_title('Salinidade')
axs[1].set_ylabel('Salinidade (PSU)')
axs[1].legend()
axs[1].grid(True)

# Plot Turbidez
for local, color in zip(locais, colors):
    local_data = dados[dados['Local'] == local]
    axs[2].plot(local_data.index, local_data['Turbidez'], linestyle='-', label=local, color=color)
    add_highlight_points(axs[2], local_data, 'Turbidez')
axs[2].set_title('Turbidez')
axs[2].set_ylabel('Turbidez (NTU)')
axs[2].set_xlabel('Tempo')
axs[2].legend()
axs[2].grid(True)

# Adicionando legenda para Levantamentos Aéreos
for ax in axs:
    ax.scatter([], [], s=100, label='Levantamentos Aéreos')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(MonthLocator())  # Locator para mostrar os marcadores de mês em mês

# Ajustando o layout
plt.tight_layout()
plt.subplots_adjust(top=0.95)

# Show the plot
plt.show()
```

```python
# para outliers ->>> cada caso deve ser analisado separadamente!!!!!
# é muito comum usar o critério como 3x o valor o desvio padrão da média
```

```python
# calcular o Z-score
z_scores = np.abs((dados['Salinidade'] - dados['Salinidade'].mean())/dados['Salinidade'].std())
```

```python
# limite para identificação de outliers ->> ex.: Z-score > 3
limiar = 3
```

```python
# agora substituição do outlier por NaN
dados['Salinidade_sem_outlier'] = dados['Salinidade']
dados.loc[z_scores > limiar, 'Salinidade_sem_outlier'] = np.nan
```

```python
# o exemplo foi usado para salinidade, que aparentemente tá limpo, mas pode ser aplicado em qq dado
```

```python
# após remoção dos valores, a série pode ser reconstruída através de interpolação
# aqui usado a interpolação cúbida, mas podem ser testadas outras!!!!!!!!
dados['Salinidade_interpolacao_cubica'] = dados['Salinidade_sem_outlier'].interpolate(method = 'cubic')
```

```python
# gráfico das duas séries (original e interpolada.. que nesse exemplo n tem diferença!)
```

```python
plt.figure(figsize = (12, 6))
plt.plot(dados.index, dados['Salinidade'], label = 'Dado original - Salinidade', color = 'b')
plt.plot(dados.index, dados['Salinidade_interpolacao_cubica'], label = 'Dado interpolado - Salinidade', color = 'r')
plt.title('Série original e interpolada de Salinidade', fontsize = 14, fontweight = 'bold')
plt.grid()
plt.ylabel('PSU', fontsize = 12), plt.xlabel('Tempo', fontsize = 12)
plt.legend()
#plt.savefig('qualquernome.png', dpi = 400) <<<- descomentar para salvar a imagem ->>> qnto maior o dpi melhor a qualidade
plt.show()
```
